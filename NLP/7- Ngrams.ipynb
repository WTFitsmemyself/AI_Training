```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sentence = 'Artificial Intelligence (AI) refers to the ability of machines and computer systems to perform tasks that traditionally require human intelligence. These tasks include problem-solving, learning, reasoning, decision-making, and even understanding natural language. AI is not a single technology but rather a broad field that combines computer science, mathematics, data science, and engineering.'
```

```python
Tokens = sentence.split()
```

```python
unigrams = (pd.Series(nltk.ngrams(Tokens, 1)).value_counts())
print(unigrams[:10])
```

```python
unigrams[:10].sort_values().plot.barh(color= "lightsalmon", width= 0.7, figsize=(12,8))
plt.title("Unigram Ngrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(Tokens, 2)).value_counts())
print(bigrams[:10])
```

```python
trigrams = (pd.Series(nltk.ngrams(Tokens, 3)).value_counts())
print(trigrams[:10])
```
